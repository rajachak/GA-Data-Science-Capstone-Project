```python
#calculate percentages
```

```python
pct_counts = raw_counts.copy()
```

```python
#change ingredient counts to ingredient percentages
for i, row in pct_counts.iterrows():
    #print(i, row['ingredient'])
    for key, val in cuisine_counts.items():
        #key - the cuisine
        #val - the total number of recipes of that cuisine
        #row[key] - the number of recipes of that cuisine with the ingredient
        #row[key] / val - perecentage of recipes with the ingredient 
        #print (key, val, row[key], row[key]/val)  
        percentage = row[key]/val
        pct_counts.loc[i, key] = percentage
```

# Similarity of cuisines

Try using cosine similarity to find most similar cuisines

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
pr2 = prevalence_ratios.T.reset_index().iloc[1:]
pr2 = pr2.set_index('cuisine')
pr2 = pr2.iloc[0:20,:]
```

```python
#pr2
```

```python
cuisine_sim = pd.DataFrame(
    cosine_similarity(pr2),
    columns=pr2.index,
    index=pr2.index
)
```

```python
cuisine_sim
```

```python
for cuisine in cuisine_counts.keys():
    print(cuisine_sim[cuisine].sort_values(ascending=False).head(10))
```
